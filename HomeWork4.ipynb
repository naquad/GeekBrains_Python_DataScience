```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('Linear regression R2 score:', r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(max_depth = 12, max_features = 5, n_estimators = 1000, random_state=42)
clf.fit(X_train, y_train.values[:, 0])
tree_y_pred = clf.predict(X_test)
```

```python
print('Random forest regressor R2 score:', r2_score(y_test, tree_y_pred))
```

```python
lr_score = r2_score(y_test, y_pred)
tree_score = r2_score(y_test, tree_y_pred)

if lr_score > tree_score:
    print('LinearRegression prediction is more accurate')
elif tree_score > lr_score:
    print('RandomForestRegressor prediction is more accurate')
else:
    print('LinearRegression and RandomForestRegressor yielded equal results. Something is wrong here :/')
```

```python
ordered = np.argsort(-clf.feature_importances_)

print('-' * 50)
print('Total importance:', clf.feature_importances_.sum())
print('Top 2:', ', '.join(X_train.columns[ordered[:2]].values.astype(str)))
print('-' * 50)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

plt.barh(X_train.columns[ordered], clf.feature_importances_[ordered])
plt.xlabel('Weight')
plt.ylabel('Feature');
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

df = pd.read_csv('./creditcard.csv')
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.drop('Class', axis=1)
y = pd.Series(df['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]

clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100), 
                   param_grid=parameters,
                   scoring='roc_auc',
                   cv=3)

clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
from sklearn.metrics import roc_auc_score

def predict_proba(clf, X, y):
    proba = clf.predict_proba(X)
    return roc_auc_score(y, proba[:, 1])

roc_auc_train = predict_proba(clf, X_train, y_train)
roc_auc_test = predict_proba(clf, X_test, y_test)

print('ROC AUC for train data:', roc_auc_train)
print('ROC AUC for test data:', roc_auc_test)
```
